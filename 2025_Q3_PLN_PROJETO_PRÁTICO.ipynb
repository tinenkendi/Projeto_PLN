```
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage
from IPython.display import display, Markdown
import os
import getpass
import requests
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableSequence
```

```
#@title Definindo a chave das APIs
os.environ["GROQ_API_KEY"] = getpass.getpass()
```

```
os.environ["GNEWS_API_KEY"] = getpass.getpass()
```

```
#@title Utilizando a API GNews para coletar notícias

def buscar_noticias(tema, max_noticias=5):
    url = f"https://gnews.io/api/v4/search?q={tema}&lang=pt&max={max_noticias}&apikey={os.environ['GNEWS_API_KEY']}"
    r = requests.get(url)
    data = r.json()
    return data.get("articles", [])
```

```
noticias = buscar_noticias("emprego", max_noticias=5)
noticias
```

```
#@title Aplicando as técnicas de PLN

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.0
)
```

```
#@title Sumarização

prompt_sumario = PromptTemplate(
    input_variables=["texto"],
    template="""
    Resuma o seguinte texto em uma frase:

    {texto}
    """
)

sumario_chain = (
    prompt_sumario
    | llm
    | StrOutputParser()
)
```

```
#@title Análise de Sentimentos

prompt_sentimento = PromptTemplate(
    input_variables=["texto"],
    template="""
    Analise o sentimento e responda apenas com as palavras (positivo, neutro ou negativo) do texto:

    {texto}
    """
)

sentimento_chain = (
    prompt_sentimento
    | llm
    | StrOutputParser()
)
```

```
#@title Detecção de Emoções

prompt_emocoes = PromptTemplate(
    input_variables=["texto"],
    template="""
    Identifique quais emoções estão presentes no texto abaixo.
    Retorne apenas com as palavras que representam as emoções mais relevantes, separadas por vírgula.

    Texto:
    {texto}
    """
)

emocoes_chain = (
    prompt_emocoes
    | llm
    | StrOutputParser()
)
```

```
#@title Extração de Palavras-chave

prompt_keywords = PromptTemplate(
    input_variables=["texto"],
    template="""
    Extraia 3 palavras-chave importantes do texto abaixo.
    Responda apenas com palavras separadas por vírgula.

    Texto:
    {texto}
    """
)

keywords_chain = (
    prompt_keywords
    | llm
    | StrOutputParser()
)
```

```
def analisar_noticia(noticia):
    texto = noticia["description"] or noticia["content"] or ""
    if not texto:
        return None

    resumo = sumario_chain.invoke({"texto": texto})
    sentimento = sentimento_chain.invoke({"texto": texto})
    emocoes = emocoes_chain.invoke({"texto": texto})
    keywords = keywords_chain.invoke({"texto": texto})

    return {
        "titulo": noticia["title"],
        "fonte": noticia["source"]["name"],
        "url": noticia["url"],
        "resumo": resumo,
        "sentimento": sentimento,
        "emocoes": emocoes,
        "palavras_chave": keywords
    }
```

```
resultados = [analisar_noticia(n) for n in noticias]
resultados
```

```
#@title Resultados Formatados

for r in resultados:
    if r is None:
        continue
    print("===============================================")
    print("TÍTULO: ", r["titulo"])
    print("FONTE: ", r["fonte"])
    print("RESUMO: ", r["resumo"])
    print("SENTIMENTO: ", r["sentimento"])
    print("EMOÇÕES:", r["emocoes"])
    print("PALAVRAS-CHAVE:", r["palavras_chave"])
    print("LINK: ", r["url"])
    print("\n")
```
